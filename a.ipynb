```python
import numpy as np
import matplotlib.pyplot as plt

#the setup of the fixed values
seed = 134
rng = np.random.default_rng(seed)
Nparticles = 15
EnergyMin = 1
EnergyChange = EnergyMin
EnergyStart = 1
Nsteps = 100

#sets up a list of all the seperate energy levels of particles for easy access later
ListOfParticleEnergies = np.full(Nparticles,EnergyMin+EnergyStart*EnergyChange)
#randomly chooses particles from the list and alters them when applicable
for x in range(0,Nsteps):
    particle1 = rng.integers(0,Nparticles)
    print(particle1)
    if ListOfParticleEnergies[particle1] != EnergyMin:
        particle2 = rng.integers(0,Nparticles)
        ListOfParticleEnergies[particle1] -= EnergyChange
        ListOfParticleEnergies[particle2] += EnergyChange

#The list of energy should have the same total energy as the start
assert np.sum(ListOfParticleEnergies) == (EnergyMin+EnergyStart*EnergyChange)*Nparticles, "There has been a change in energy! There has been an error."

# print(ListOfParticleEnergies)
```

```python


plt.hist(ListOfParticleEnergies,bins=128)
plt.yscale('log')
x = np.arange(0,30)
plt.plot(x, np.exp(-x/(1+EnergyStart))*160)
```

```python
import numpy as np
N = 15
e0 = 1
de = 1
nsteps = 100
energy = np.full(N,e0+de)

print("initial energy",energy.mean())
rng = np.random.default_rng(134)

for iteration in range(nsteps):

    particle_index = rng.integers(N)
    print(particle_index)
    
    # Check if the particle is in the ground state
    if energy[particle_index] == e0:
        continue

    # Randomly choose another particle
    other_particle_index = rng.integers(N)
    
    # Skip if the same particle is chosen
    if particle_index == other_particle_index:
        continue
    
    # Perform energy exchange
    energy[particle_index] -= de
    energy[other_particle_index] += de


print("finally energy", energy.mean())

import matplotlib.pyplot as plt
# plt.plot(energy)


plt.figure()

bins= np.arange(energy.min(),energy.max()+de,de)
plt.hist(energy,bins=bins)
plt.plot(N*np.exp(-bins/energy.mean()))
plt.yscale('log')
```
